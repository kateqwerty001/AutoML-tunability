```python
# roc_auc gained by default models

extra_trees_default_1 = pd.read_csv('RandomSearch/ExtraTreesClassifier/extra_trees_1_default.csv')
extra_trees_default_2 = pd.read_csv('RandomSearch/ExtraTreesClassifier/extra_trees_2_default.csv')
extra_trees_default_3 = pd.read_csv('RandomSearch/ExtraTreesClassifier/extra_trees_3_default.csv')
extra_trees_default_4 = pd.read_csv('RandomSearch/ExtraTreesClassifier/extra_trees_4_default.csv')

extra_trees_default_1 = extra_trees_default_1['roc_auc']
extra_trees_default_2 = extra_trees_default_2['roc_auc']
extra_trees_default_3 = extra_trees_default_3['roc_auc']
extra_trees_default_4 = extra_trees_default_4['roc_auc']

svm_default_1 = pd.read_csv('RandomSearch/SupportVectorMachine/svm_1_default.csv')
svm_default_2 = pd.read_csv('RandomSearch/SupportVectorMachine/svm_2_default.csv')
svm_default_3 = pd.read_csv('RandomSearch/SupportVectorMachine/svm_3_default.csv')
svm_default_4 = pd.read_csv('RandomSearch/SupportVectorMachine/svm_4_default.csv')

svm_default_1 = svm_default_1['roc_auc']
svm_default_2 = svm_default_2['roc_auc']
svm_default_3 = svm_default_3['roc_auc']
svm_default_4 = svm_default_4['roc_auc']

catboost_default_1 = pd.read_csv('RandomSearch/CatBoost/catboost_1_default.csv')
catboost_default_2 = pd.read_csv('RandomSearch/CatBoost/catboost_2_default.csv')
catboost_default_3 = pd.read_csv('RandomSearch/CatBoost/catboost_3_default.csv')
catboost_default_4 = pd.read_csv('RandomSearch/CatBoost/catboost_4_default.csv')

catboost_default_1 = catboost_default_1['roc_auc']
catboost_default_2 = catboost_default_2['roc_auc']
catboost_default_3 = catboost_default_3['roc_auc']
catboost_default_4 = catboost_default_4['roc_auc']

colors = ['#FFD18E', '#75A47F', '#FFD18E', '#75A47F', '#FFD18E', '#75A47F', '#FFD18E', '#75A47F']
```

# Support Vector Machine

```python
# boxplot data

data_for_plot= pd.DataFrame()
data_for_plot['roc_auc_random_1'] = svm_1_random
data_for_plot['roc_auc_bayes_1'] = svm_1_bayes
data_for_plot['roc_auc_random_2'] = svm_2_random
data_for_plot['roc_auc_bayes_2'] = svm_2_bayes
data_for_plot['roc_auc_random_3'] = svm_3_random
data_for_plot['roc_auc_bayes_3'] = svm_3_bayes
data_for_plot['roc_auc_random_4'] = svm_4_random
data_for_plot['roc_auc_bayes_4'] = svm_4_bayes


plt.figure(figsize=(9, 7))
sns.boxplot(data=data_for_plot, palette=colors, boxprops=dict(edgecolor='black'), whiskerprops=dict(color='black'), capprops=dict(color='black'), medianprops=dict(color='black'))
sns.stripplot(data=data_for_plot, jitter=True, color="black", size=2, edgecolor='black')

# default models roc_auc
plt.scatter([0], [svm_default_1], color='#fc4e2a', edgecolor='black', linewidth=2.5, s=90, zorder=10)
plt.scatter([1], [svm_default_1], color='#fc4e2a', edgecolor='black', linewidth=2.5, s=90, zorder=10)
plt.scatter([2], [svm_default_2], color='#fc4e2a', edgecolor='black', linewidth=2.5, s=90, zorder=10)
plt.scatter([3], [svm_default_2], color='#fc4e2a', edgecolor='black', linewidth=2.5, s=90, zorder=10)
plt.scatter([4], [svm_default_3], color='#fc4e2a', edgecolor='black', linewidth=2.5, s=90, zorder=10)
plt.scatter([5], [svm_default_3], color='#fc4e2a', edgecolor='black', linewidth=2.5, s=90, zorder=10)
plt.scatter([6], [svm_default_4], color='#fc4e2a', edgecolor='black', linewidth=2.5, s=90, zorder=10)
plt.scatter([7], [svm_default_4], color='#fc4e2a', edgecolor='black', linewidth=2.5, s=90, zorder=10)

plt.title('ROC AUC for SVM models (default models in red), Random Search vs Bayesian Optimization')
plt.ylabel('ROC AUC')
plt.xlabel('Dataset number', rotation=0)
plt.xticks(rotation=90, ticks=[0.5, 2.5, 4.5, 6.5], labels=['833', '44157', '1120', '45553'])

legend_handles = [Line2D([0], [0], marker='o', color='w', markerfacecolor='#FFD18E', markersize=10, label='Random Search'),
                Line2D([0], [0], marker='o', color='w', markerfacecolor='#75A47F', markersize=10, label='Bayesian Optimization')]

plt.legend(handles=legend_handles)
plt.show()
```
